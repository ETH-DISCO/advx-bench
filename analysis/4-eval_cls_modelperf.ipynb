```python
pivot_table = pd.merge(pivot_table_at_1, pivot_table_at_5, on=["model", "mask"])

# rename columns
pivot_table.columns = [col.replace("_", "-") for col in pivot_table.columns]

# turn all to strings
for col in pivot_table.columns:
    if pivot_table[col].dtype != object:
        pivot_table[col] = pivot_table[col].round(3).astype(str)

# write @1 and @5 to the same cell
for col in pivot_table.columns:
    if "change@1" in col:
        pivot_table[col] = pivot_table[col] + " (" + pivot_table[col.replace("change@1", "change@5")] + ")"

pivot_table = pivot_table.drop(columns=[col for col in pivot_table.columns if "change@5" in col])

display(pivot_table)

print(pivot_table.to_latex(index=False, float_format="%.2f"))
```
